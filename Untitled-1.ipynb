```python
import cv2
from fer import FER
from collections import deque, Counter
import numpy as np
import sounddevice as sd
import time

# =================================================
# CONFIGURATION
# =================================================
WINDOW_SIZE = 15            # facial smoothing (frames)
AUDIO_WINDOW = 10           # audio energy window
TIME_WINDOW = 10            # seconds for temporal reasoning

VALENCE_THRESHOLD = -0.3
AROUSAL_THRESHOLD = 0.3
VOICE_ENERGY_THRESHOLD = 5

# =================================================
# INITIALIZE CAMERA & FER
# =================================================
detector = FER(mtcnn=True)
cap = cv2.VideoCapture(0)

emotion_window = deque(maxlen=WINDOW_SIZE)

# =================================================
# AUDIO (VOICE FUSION)
# =================================================
energy_window = deque(maxlen=AUDIO_WINDOW)

def audio_callback(indata, frames, time_info, status):
    volume_norm = np.linalg.norm(indata) * 10
    energy_window.append(volume_norm)

def estimate_voice_fatigue():
    if len(energy_window) < AUDIO_WINDOW:
        return False
    avg_energy = sum(energy_window) / len(energy_window)
    return avg_energy < VOICE_ENERGY_THRESHOLD

audio_stream = sd.InputStream(
    callback=audio_callback,
    channels=1,
    samplerate=16000
)
audio_stream.start()

# =================================================
# TEMPORAL WINDOW (10s reasoning)
# =================================================
temporal_buffer = []
last_window_time = time.time()
previous_summary = None

# =================================================
# MAIN LOOP
# =================================================
print("Emotion-Aware Study Companion started")
print("Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    emotions = detector.detect_emotions(frame)

    if emotions:
        # -----------------------------------------
        # Facial emotions (smoothed)
        # -----------------------------------------
        emotions_dict = emotions[0]["emotions"]
        emotion_window.append(emotions_dict)

        avg_emotions = {
            e: sum(f[e] for f in emotion_window) / len(emotion_window)
            for e in emotions_dict
        }

        # -----------------------------------------
        # Valence & Arousal
        # -----------------------------------------
        valence = (
            avg_emotions["happy"]
            - avg_emotions["sad"]
            - avg_emotions["angry"]
        )

        arousal = (
            avg_emotions["angry"]
            + avg_emotions["happy"]
            - avg_emotions["neutral"]
        )

        # -----------------------------------------
        # Voice fusion
        # -----------------------------------------
        voice_fatigue = estimate_voice_fatigue()

        # -----------------------------------------
        # Instant affective state
        # -----------------------------------------
        if voice_fatigue and valence < -0.2:
            affective_state = "Fatigue (face + voice)"
            suggestion = "Short break recommended"

        elif valence < VALENCE_THRESHOLD and arousal < AROUSAL_THRESHOLD:
            affective_state = "Low mood / fatigue"
            suggestion = "Consider a short break"

        elif valence < VALENCE_THRESHOLD and arousal >= AROUSAL_THRESHOLD:
            affective_state = "Frustration / stress"
            suggestion = "Slow down or breathe"

        else:
            affective_state = "Neutral / engaged"
            suggestion = "Keep going"

        # -----------------------------------------
        # Store frame-level data for 10s window
        # -----------------------------------------
        temporal_buffer.append({
            "timestamp": time.time(),
            "valence": valence,
            "arousal": arousal,
            "voice_fatigue": voice_fatigue,
            "state": affective_state
        })

        # -----------------------------------------
        # DISPLAY (current state)
        # -----------------------------------------
        cv2.putText(frame, f"Valence: {valence:.2f} | Arousal: {arousal:.2f}",
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        cv2.putText(frame, f"State: {affective_state}",
                    (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2)

        cv2.putText(frame, f"Suggestion: {suggestion}",
                    (20, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 200, 200), 2)

        cv2.putText(frame,
                    f"Voice: {'Low energy' if voice_fatigue else 'Normal'}",
                    (20, 160), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 200, 255), 2)

    else:
        cv2.putText(frame, "No face detected",
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # =================================================
    # 10-SECOND TEMPORAL SUMMARY
    # =================================================
    current_time = time.time()

    if current_time - last_window_time >= TIME_WINDOW and temporal_buffer:
        valences = [x["valence"] for x in temporal_buffer]
        arousals = [x["arousal"] for x in temporal_buffer]
        voice_flags = [x["voice_fatigue"] for x in temporal_buffer]
        states = [x["state"] for x in temporal_buffer]

        avg_valence = sum(valences) / len(valences)
        avg_arousal = sum(arousals) / len(arousals)
        voice_ratio = sum(voice_flags) / len(voice_flags)
        dominant_state = Counter(states).most_common(1)[0][0]

        if avg_valence < -0.3 and voice_ratio > 0.6:
            conclusion = "Sustained fatigue – break recommended"
        elif avg_valence < -0.3:
            conclusion = "Negative mood – slow down"
        else:
            conclusion = "Stable / engaged"

        previous_summary = {
            "avg_valence": avg_valence,
            "avg_arousal": avg_arousal,
            "voice_ratio": voice_ratio,
            "state": dominant_state,
            "conclusion": conclusion
        }

        # Console output (for demo explanation)
        print("\n===== 10s WINDOW SUMMARY =====")
        print(f"Avg valence: {avg_valence:.2f}")
        print(f"Avg arousal: {avg_arousal:.2f}")
        print(f"Voice fatigue ratio: {voice_ratio:.2f}")
        print(f"Dominant state: {dominant_state}")
        print(f"Conclusion: {conclusion}")

        temporal_buffer.clear()
        last_window_time = current_time

    # Overlay last 10s conclusion
    if previous_summary:
        cv2.putText(frame,
                    f"Last 10s: {previous_summary['conclusion']}",
                    (20, 200), cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (0, 200, 255), 2)

    cv2.imshow("Emotion-Aware Study Companion", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# =================================================
# CLEANUP
# =================================================
audio_stream.stop()
cap.release()
cv2.destroyAllWindows()
```
